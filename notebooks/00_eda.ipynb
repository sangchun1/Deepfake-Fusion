## 2. 헬퍼 함수 정의

```python
def get_image_paths(root_dir, split, label):
    exts = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}
    paths = []
    for p in sorted(root_dir.rglob("*")):
        if p.suffix.lower() in exts:
            paths.append({
                "filepath": str(p),
                "filename": p.name,
                "split": split,
                "label": label
            })
    return paths
```

```python
def image_basic_stats(img):
    arr = np.array(img).astype(np.float32) / 255.0
    
    gray = cv2.cvtColor((arr * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    
    brightness = gray.mean()
    contrast = gray.std()
    rgb_mean = arr.mean(axis=(0, 1))
    rgb_std = arr.std(axis=(0, 1))
    
    # Shannon entropy
    hist, _ = np.histogram(gray.flatten(), bins=256, range=(0, 1), density=True)
    hist = hist + 1e-12
    entropy = -(hist * np.log2(hist)).sum()
    
    # Sharpness (Laplacian variance)
    lap_var = cv2.Laplacian((gray * 255).astype(np.uint8), cv2.CV_64F).var()
    
    # Edge density
    edges = cv2.Canny((gray * 255).astype(np.uint8), 100, 200)
    edge_density = (edges > 0).mean()
    
    return {
        "width": img.size[0],
        "height": img.size[1],
        "brightness": brightness,
        "contrast": contrast,
        "r_mean": rgb_mean[0],
        "g_mean": rgb_mean[1],
        "b_mean": rgb_mean[2],
        "r_std": rgb_std[0],
        "g_std": rgb_std[1],
        "b_std": rgb_std[2],
        "entropy": entropy,
        "lap_var": lap_var,
        "edge_density": edge_density
    }
```

```python
def compute_fft_features(img):
    gray = np.array(img.convert("L")).astype(np.float32) / 255.0
    
    f = np.fft.fft2(gray)
    fshift = np.fft.fftshift(f)
    magnitude = np.log1p(np.abs(fshift))
    
    h, w = magnitude.shape
    cy, cx = h // 2, w // 2
    y, x = np.indices((h, w))
    r = np.sqrt((x - cx)**2 + (y - cy)**2)
    
    r = r.astype(np.int32)
    radial_mean = np.bincount(r.ravel(), magnitude.ravel()) / np.bincount(r.ravel())
    
    # 중심부(저주파) / 바깥쪽(고주파) 에너지 비율
    max_r = r.max()
    low_mask = r <= max_r * 0.25
    high_mask = r >= max_r * 0.60
    
    low_energy = magnitude[low_mask].mean()
    high_energy = magnitude[high_mask].mean()
    
    return {
        "fft_magnitude": magnitude,
        "radial_profile": radial_mean,
        "low_freq_energy": low_energy,
        "high_freq_energy": high_energy,
        "high_low_ratio": high_energy / (low_energy + 1e-8)
    }
```

```python
def show_image_grid(df, n=16, title="Sample Images"):
    sample_df = df.sample(min(n, len(df)), random_state=SEED).reset_index(drop=True)
    
    cols = 4
    rows = int(np.ceil(len(sample_df) / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(12, 3 * rows))
    axes = np.array(axes).reshape(-1)
    
    for ax in axes:
        ax.axis("off")
    
    for i, row in sample_df.iterrows():
        img = Image.open(row["filepath"]).convert("RGB")
        axes[i].imshow(img)
        axes[i].set_title(f'{row["split"]} | {row["label"]}', fontsize=10)
        axes[i].axis("off")
    
    fig.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()
```

## 8. Frequency-Domain Analysis

공간 도메인에서는 잘 안 보이는 차이가 주파수 영역에서 나타나는지 확인한다.

```python
fft_records = []
fft_maps_real = []
fft_maps_fake = []

sample_for_fft = df.groupby(["split", "label"], group_keys=False).apply(
    lambda x: x.sample(min(2000, len(x)), random_state=SEED)
).reset_index(drop=True)

for _, row in tqdm(sample_for_fft.iterrows(), total=len(sample_for_fft)):
    img = Image.open(row["filepath"]).convert("RGB")
    f = compute_fft_features(img)
    
    fft_records.append({
        "filepath": row["filepath"],
        "label": row["label"],
        "split": row["split"],
        "low_freq_energy": f["low_freq_energy"],
        "high_freq_energy": f["high_freq_energy"],
        "high_low_ratio": f["high_low_ratio"],
        "radial_profile": f["radial_profile"]
    })
    
    if row["label"] == "REAL":
        fft_maps_real.append(f["fft_magnitude"])
    else:
        fft_maps_fake.append(f["fft_magnitude"])

fft_df = pd.DataFrame(fft_records)
```

## 9. Embedding-Space Exploration

간단한 특징 벡터로 차원을 축소해 REAL / FAKE가 어느 정도 분리되는지 살펴본다.

```python
from torchvision import models, transforms
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
embed_df = df.groupby(["split", "label"], group_keys=False).apply(
    lambda x: x.sample(min(1000, len(x)), random_state=SEED)
).reset_index(drop=True)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

class ImagePathDataset(Dataset):
    def __init__(self, paths, transform=None):
        self.paths = paths
        self.transform = transform
        
    def __len__(self):
        return len(self.paths)
    
    def __getitem__(self, idx):
        path = self.paths[idx]
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, path

dataset = ImagePathDataset(embed_df["filepath"].tolist(), transform=transform)
loader = DataLoader(dataset, batch_size=64, shuffle=False, num_workers=2)
```

```python
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
model.fc = torch.nn.Identity()
model = model.to(device).eval()
```

```python
embeddings = []
paths = []

with torch.no_grad():
    for imgs, batch_paths in tqdm(loader):
        imgs = imgs.to(device)
        feats = model(imgs)
        embeddings.append(feats.cpu().numpy())
        paths.extend(batch_paths)

embeddings = np.concatenate(embeddings, axis=0)
embeddings.shape
```

```python
embed_vis_df = embed_df.copy().reset_index(drop=True)

pca = PCA(n_components=2, random_state=SEED)
pca_2d = pca.fit_transform(embeddings)

embed_vis_df["pca1"] = pca_2d[:, 0]
embed_vis_df["pca2"] = pca_2d[:, 1]
```

```python
plt.figure(figsize=(8, 6))

for label in ["REAL", "FAKE"]:
    sub = embed_vis_df[embed_vis_df["label"] == label]
    plt.scatter(sub["pca1"], sub["pca2"], s=10, alpha=0.5, label=label)

plt.title("PCA of Image Embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()
```

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=SEED, init="pca")
tsne_2d = tsne.fit_transform(embeddings[:2000])

tsne_df = embed_vis_df.iloc[:2000].copy()
tsne_df["x"] = tsne_2d[:, 0]
tsne_df["y"] = tsne_2d[:, 1]

plt.figure(figsize=(8, 6))
for label in ["REAL", "FAKE"]:
    sub = tsne_df[tsne_df["label"] == label]
    plt.scatter(sub["x"], sub["y"], s=10, alpha=0.5, label=label)

plt.title("t-SNE of Image Embeddings")
plt.legend()
plt.show()
```

## 12. Findings for Modeling

### 요약
- 데이터 구조는 train/test, REAL/FAKE 이진 분류 형태로 정리되어 있다.
- 이미지 해상도와 채널 수가 대부분 일관적이다.
- REAL / FAKE 간 밝기, 엔트로피, sharpness, edge density 등에서 차이가 존재할 수 있다.
- 주파수 분석에서 특정 대역 차이가 관찰되면 frequency branch 도입의 근거가 된다.
- 임베딩 공간에서 일부 군집 분리가 보이면 baseline CNN만으로도 성능이 나올 수 있지만,
  동시에 artifact 중심 학습 가능성도 고려해야 한다.
- 따라서 이후 실험은
  1) spatial baseline,
  2) frequency-only baseline,

## 2. 헬퍼 함수 정의

CIFAKE에서 쓴 `safe_open_image`, `image_basic_stats`, `compute_fft_features`, `show_image_grid`는 그대로 재사용해도 된다.

```python
def infer_face130k_meta(path_str):
    path = Path(path_str)
    parts = [p.lower() for p in path.parts]
    joined = "/".join(parts)

    # split 추론
    split = "all"
    for cand in ["train", "val", "valid", "validation", "test"]:
        if cand in parts:
            split = "val" if cand in ["val", "valid", "validation"] else cand
            break

    # label 추론
    label = "UNKNOWN"
    if any(x in parts for x in ["real", "ffhq"]):
        label = "REAL"
    elif any(x in parts for x in ["fake", "synthetic", "ai", "generated"]):
        label = "FAKE"

    # source 추론
    source = "unknown"
    source_candidates = {
        "ffhq": "FFHQ",
        "flux1.dev": "FLUX1_DEV",
        "flux1.pro": "FLUX1_PRO",
        "flux": "FLUX",
        "sdxl": "SDXL",
        "stable-diffusion-xl": "SDXL"
    }
    for k, v in source_candidates.items():
        if k in joined:
            source = v
            break

    return split, label, source
```

```python
def get_face130k_image_paths(root_dir):
    exts = {".jpg", ".jpeg", ".png", ".bmp", ".webp"}
    records = []

    for p in sorted(root_dir.rglob("*")):
        if p.suffix.lower() in exts:
            split, label, source = infer_face130k_meta(str(p))
            records.append({
                "filepath": str(p),
                "filename": p.name,
                "split": split,
                "label": label,
                "source": source
            })
    return records
```

```python
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

def detect_face_stats(img):
    arr = np.array(img.convert("RGB"))
    gray = cv2.cvtColor(arr, cv2.COLOR_RGB2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    h, w = gray.shape
    result = {
        "n_faces": len(faces),
        "face_found": int(len(faces) > 0),
        "face_area_ratio": np.nan,
        "face_center_offset": np.nan,
        "face_w_ratio": np.nan,
        "face_h_ratio": np.nan
    }

    if len(faces) > 0:
        # 가장 큰 얼굴 1개 기준
        areas = [fw * fh for (_, _, fw, fh) in faces]
        idx = int(np.argmax(areas))
        x, y, fw, fh = faces[idx]

        face_area = fw * fh
        img_area = w * h

        face_cx = x + fw / 2
        face_cy = y + fh / 2
        img_cx = w / 2
        img_cy = h / 2

        center_offset = np.sqrt((face_cx - img_cx)**2 + (face_cy - img_cy)**2)
        center_offset /= np.sqrt(img_cx**2 + img_cy**2)

        result.update({
            "face_area_ratio": face_area / img_area,
            "face_center_offset": center_offset,
            "face_w_ratio": fw / w,
            "face_h_ratio": fh / h
        })

    return result
```

```python
def symmetry_score(img):
    gray = np.array(img.convert("L")).astype(np.float32) / 255.0
    h, w = gray.shape

    left = gray[:, :w // 2]
    right = gray[:, w - (w // 2):]
    right = np.fliplr(right)

    min_w = min(left.shape[1], right.shape[1])
    left = left[:, :min_w]
    right = right[:, :min_w]

    return np.mean(np.abs(left - right))
```

```python
def jpeg_blockiness_score(img):
    gray = np.array(img.convert("L")).astype(np.float32)
    h, w = gray.shape

    if h < 16 or w < 16:
        return np.nan

    # 8x8 block 경계 차이의 단순 proxy
    vertical_edges = []
    for c in range(8, w, 8):
        vertical_edges.append(np.abs(gray[:, c] - gray[:, c - 1]).mean())

    horizontal_edges = []
    for r in range(8, h, 8):
        horizontal_edges.append(np.abs(gray[r, :] - gray[r - 1, :]).mean())

    return float(np.mean(vertical_edges + horizontal_edges)) if (vertical_edges or horizontal_edges) else np.nan
```

## 11. Embedding-Space Exploration

pretrained encoder 기반 임베딩으로 REAL / FAKE 및 source별 군집을 확인한다.

```python
from torchvision import models, transforms
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
embed_df = face_df.groupby("label", group_keys=False).apply(
    lambda x: x.sample(min(1000, len(x)), random_state=SEED)
).reset_index(drop=True)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

class ImagePathDataset(Dataset):
    def __init__(self, paths, transform=None):
        self.paths = paths
        self.transform = transform

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        path = self.paths[idx]
        img = Image.open(path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, path

dataset = ImagePathDataset(embed_df["filepath"].tolist(), transform=transform)
loader = DataLoader(dataset, batch_size=64, shuffle=False, num_workers=2)
```

```python
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
model.fc = torch.nn.Identity()
model = model.to(device).eval()
```

```python
embeddings = []
paths = []

with torch.no_grad():
    for imgs, batch_paths in tqdm(loader):
        imgs = imgs.to(device)
        feats = model(imgs)
        embeddings.append(feats.cpu().numpy())
        paths.extend(batch_paths)

embeddings = np.concatenate(embeddings, axis=0)
embeddings.shape
```

## 14. Findings for Modeling

### 요약 템플릿
- 데이터 구조는 [train/test 여부] / [REAL-FAKE 비율] 형태로 구성되어 있다.
- REAL / FAKE 간 해상도 및 파일 크기 차이가 [존재/미미]하다.
- 얼굴 영역의 크기/정렬 패턴 차이가 [존재/미미]하다.
- 텍스처 및 sharpness / entropy / symmetry / blockiness 차이가 [관찰됨/약함]하다.
- 주파수 영역에서 [특정 대역 차이]가 관찰된다 / 관찰되지 않는다.
- source별 분포 차이가 [큼/작음]하다.

### 모델링 시사점
1. resize / crop / compression을 통제하지 않으면 shortcut learning 위험이 있다.
2. baseline CNN과 함께 frequency branch를 두는 실험이 타당하다.
3. 얼굴 alignment 여부를 실험 변수로 둘 가치가 있다.
4. source별 split 또는 cross-source validation을 고려할 필요가 있다.
5. 오류 분석에서는 눈, 치아, 머리카락, 배경 영역을 우선 확인한다.
